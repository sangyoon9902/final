```python
# scrape_kspo_prescription_videos.py
import re, time, sys, math
from dataclasses import dataclass, asdict
from typing import List, Optional
import pandas as pd
from datetime import datetime

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

LIST_URL = "https://nfa.kspo.or.kr/classroom/program/selectPrescriptionMovieList.kspo"

# í˜ì´ì§€ê°€ ì¢…ì¢… ëŠë¦¬ê²Œ ë¡œë“œë  ìˆ˜ ìˆì–´ ì—¬ìœ ìˆê²Œ
WAIT_SEC = 12
PAGE_PAUSE = 1.2  # í˜ì´ì§€ ì „í™˜ê°„ ì¿¨ë‹¤ìš´
ITEM_PAUSE = 0.05  # ì¹´ë“œ íŒŒì‹±ê°„ ì§§ì€ ì¿¨ë‹¤ìš´
PARTIAL_SAVE_EVERY = 50  # N í˜ì´ì§€ë§ˆë‹¤ ì„ì‹œ ì €ì¥

YOUTUBE_ID_RE = re.compile(r"/vi/([A-Za-z0-9_-]{6,})/")


@dataclass
class VideoRow:
    title: str
    fitness_category: Optional[str]
    tool: Optional[str]
    body_part: Optional[str]
    target: Optional[str]
    disease: Optional[str]
    youtube_id: Optional[str]
    youtube_url: Optional[str]
    thumb_url: Optional[str]
    page_no: int
    rank_on_page: int


def extract_text_after_prefix(lines, prefix):
    for ln in lines:
        t = re.sub(r"\s+", " ", ln).strip()
        if t.startswith(prefix):
            return re.split(r"\s*:\s*", t, maxsplit=1)[-1].strip() or None
    return None


def human_seconds(sec: float) -> str:
    m, s = divmod(int(sec), 60)
    h, m = divmod(m, 60)
    if h:
        return f"{h}h {m}m {s}s"
    if m:
        return f"{m}m {s}s"
    return f"{s}s"


# -------------------------
# ê³µí†µ ìœ í‹¸
# -------------------------
def get_total_count(driver) -> Optional[int]:
    """'ì´ N ê°œ' ë¬¸êµ¬ì—ì„œ N ì¶”ì¶œ"""
    try:
        el = driver.find_element(
            By.XPATH,
            "//*[contains(normalize-space(),'ì´') and contains(normalize-space(),'ê°œ') and contains(normalize-space(),'ê²Œì‹œ')]",
        )
        txt = (el.text or "").replace(",", "")
        m = re.search(r"ì´\s*([0-9]+)\s*ê°œ", txt)
        if m:
            return int(m.group(1))
    except Exception:
        pass
    return None


def scroll_into_view_and_click(driver, elem):
    driver.execute_script("arguments[0].scrollIntoView({block:'center'});", elem)
    WebDriverWait(driver, WAIT_SEC).until(EC.element_to_be_clickable(elem))
    elem.click()


def collect_thumb_imgs(driver):
    return driver.find_elements(
        By.XPATH,
        "//img[contains(@alt,'ìš´ë™ì²˜ë°© ë™ì˜ìƒ') or contains(@src,'img.youtube.com/vi/')]",
    )


def collect_cards_from_thumbs(driver):
    thumbs = collect_thumb_imgs(driver)
    cards = []
    for img in thumbs:
        card = None
        try:
            card = img.find_element(By.XPATH, "ancestor::li[1]")
        except:
            try:
                card = img.find_element(By.XPATH, "ancestor::div[1]")
            except:
                pass
        if card:
            cards.append(card)
    # dedupe by rect
    uniq = {}
    for el in cards:
        try:
            r = el.rect
            key = (r.get("x"), r.get("y"), r.get("width"), r.get("height"))
        except:
            key = (el.text or "")[:80]
        uniq[key] = el
    return list(uniq.values())


def get_page_signature(driver) -> str:
    """
    í˜„ì¬ í˜ì´ì§€ì˜ ì¸ë„¤ì¼ src 3ê°œ ì •ë„ë¥¼ í•©ì³ ì„œëª… ë¬¸ìì—´ë¡œ ìƒì„±.
    (ì²« ì¹´ë“œê°€ ê°™ì•„ë„ ì „ì²´ ì„œëª…ì´ ë‹¬ë¼ì§ˆ í™•ë¥ ì´ ë†’ìŒ)
    """
    thumbs = collect_thumb_imgs(driver)[:3]
    sigs = []
    for t in thumbs:
        try:
            sigs.append(t.get_attribute("src") or "")
        except:
            sigs.append("")
    return "|".join(sigs)


# -------------------------
# ë„¤ë¹„ê²Œì´ì…˜
# -------------------------
def call_js_pagination(driver, target_page) -> bool:
    """
    ì‚¬ì´íŠ¸ì— ì¡´ì¬í•  ìˆ˜ ìˆëŠ” í˜ì´ì§• JS í•¨ìˆ˜ë¥¼ ìˆœì°¨ ì‹œë„.
    ì„±ê³µ/ì‹¤íŒ¨ë§Œ ë°˜í™˜.
    """
    candidates = [
        "if (typeof fn_search==='function') { fn_search(arguments[0]); return true; } return false;",
        "if (typeof goPage==='function') { goPage(arguments[0]); return true; } return false;",
        "if (typeof movePage==='function') { movePage(arguments[0]); return true; } return false;",
        "if (typeof fnSearch==='function') { fnSearch(arguments[0]); return true; } return false;",
        "if (typeof paging==='function') { paging(arguments[0]); return true; } return false;",
    ]
    for js in candidates:
        try:
            ok = driver.execute_script(js, int(target_page))
            if ok:
                return True
        except Exception:
            pass
    # hidden input ë°©ì‹(ì˜ˆ: form ì•ˆ pageIndex)ë„ ì‹œë„
    try:
        js = """
        var f = document.querySelector('form');
        if (!f) return false;
        var ip = f.querySelector('input[name=pageIndex],input[name=page],input[name=currentPage]');
        if (!ip) {
            ip = document.createElement('input'); ip.type='hidden'; ip.name='pageIndex'; f.appendChild(ip);
        }
        ip.value = arguments[0];
        if (typeof f.submit === 'function') { f.submit(); return true; }
        return false;
        """
        ok = driver.execute_script(js, int(target_page))
        if ok:
            return True
    except Exception:
        pass
    return False


def goto_page(driver, wait, target_page, page_pause=1.0, prev_sig=None) -> bool:
    """
    1) ìˆ«ì ë§í¬ í´ë¦­
    2) ì‹¤íŒ¨ ì‹œ JS í˜ì´ì§• í•¨ìˆ˜ í˜¸ì¶œ
    3) ê·¸ë˜ë„ ì‹¤íŒ¨ë©´ 'ë‹¤ìŒ' ì—¬ëŸ¬ ë²ˆ í´ë¦­í•´ì„œ ëª©í‘œê¹Œì§€
    í˜ì´ì§€ ë³€ê²½ ì—¬ë¶€ëŠ” ì„œëª… ë¹„êµë¡œ ê²€ì¦.
    """
    # ëª©í‘œ í˜ì´ì§€ê°€ 2 ì´ìƒì¼ ë•Œë§Œ ì´ë™ ì‹œë„
    if target_page <= 1:
        return True

    before_sig = prev_sig or get_page_signature(driver)

    # 1) ìˆ«ì ë§í¬ í´ë¦­ ì‹œë„ (í˜„ì¬ ë·°ì—ì„œ ë³´ì´ëŠ” ê²½ìš°ë§Œ)
    try:
        link = driver.find_element(
            By.XPATH, f"//a[normalize-space(text())='{target_page}']"
        )
        scroll_into_view_and_click(driver, link)
        time.sleep(page_pause)
        wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
        after_sig = get_page_signature(driver)
        if after_sig and after_sig != before_sig:
            return True
    except Exception as e:
        print(f"âš ï¸ í˜ì´ì§€ {target_page} ìˆ«ì ë§í¬ í´ë¦­ ì‹¤íŒ¨: {e}")

    # 2) JS í˜ì´ì§• í•¨ìˆ˜ í˜¸ì¶œ
    try:
        if call_js_pagination(driver, target_page):
            time.sleep(page_pause)
            wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
            after_sig = get_page_signature(driver)
            if after_sig and after_sig != before_sig:
                return True
    except Exception as e:
        print(f"âš ï¸ JS í˜ì´ì§• í˜¸ì¶œ ì‹¤íŒ¨: {e}")

    # 3) 'ë‹¤ìŒ'ì„ ë°˜ë³µ í´ë¦­í•˜ì—¬ ëª©í‘œê¹Œì§€ ì „ì§„
    try:
        # í˜„ì¬ ì‹œê·¸ë‹ˆì²˜
        cur_sig = before_sig
        # í•œ ë²ˆì— ë„ˆë¬´ ë§ì´ ì‹œë„í•˜ì§€ ì•Šë„ë¡ ì•ˆì „ ìƒí•œ
        tries = 0
        while tries < 8:
            next_candidates = driver.find_elements(By.XPATH, "//a[contains(.,'ë‹¤ìŒ')]")
            if not next_candidates:
                next_candidates = driver.find_elements(By.CSS_SELECTOR, ".next, a.next")
            if not next_candidates:
                print("âš ï¸ 'ë‹¤ìŒ' ë§í¬ë¥¼ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")
                break

            scroll_into_view_and_click(driver, next_candidates[0])
            try:
                WebDriverWait(driver, WAIT_SEC).until(
                    EC.presence_of_element_located((By.TAG_NAME, "body"))
                )
            except:
                pass
            time.sleep(page_pause)

            new_sig = get_page_signature(driver)
            if new_sig and new_sig != cur_sig:
                cur_sig = new_sig
                # ëª©í‘œ í˜ì´ì§€ì— ë„ë‹¬í–ˆëŠ”ì§€ëŠ” ìˆ«ì ë§í¬ê°€ ë³´ì´ëŠ”ì§€ë¡œ í•œ ë²ˆ ë” ì ê²€
                # (ë³´ì´ì§€ ì•Šì•„ë„ ìµœì¢…ì ìœ¼ë¡œëŠ” next ë°˜ë³µìœ¼ë¡œ ëª¨ë“  ì¹´ë“œ ìˆ˜ì§‘ ê°€ëŠ¥)
                tries += 1
                # ê³„ì† ì „ì§„ (ëª©í‘œ ì •í™• ë„ë‹¬ ì²´í¬ê°€ ì–´ë ¤ìš´ êµ¬ì¡°ë¼ë©´ ëª¨ë“  í˜ì´ì§€ë¥¼ ìˆœì°¨ ì „ì§„ìœ¼ë¡œ ì²˜ë¦¬)
                # ì—¬ê¸°ì„  after_sig ë³€í™”ë§Œ í™•ì¸í•˜ê³  ë£¨í”„ë¥¼ ë¹ ì ¸ë‚˜ê°
                return True
            else:
                tries += 1

    except Exception as e:
        print(f"âš ï¸ 'ë‹¤ìŒ' ë°˜ë³µ í´ë¦­ ì¤‘ ì˜¤ë¥˜: {e}")

    return False


# -------------------------
# ë©”ì¸ ìŠ¤í¬ë ˆì´í¼
# -------------------------
def scrape() -> List[VideoRow]:
    t0 = time.time()

    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1280,1800")
    options.add_argument(
        "user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    )

    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()), options=options
    )
    wait = WebDriverWait(driver, WAIT_SEC)

    print(f"ğŸš€ ì‹œì‘: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"ëª©ë¡ URL: {LIST_URL}")

    try:
        driver.get(LIST_URL)
        wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

        # === ì²« í˜ì´ì§€ì—ì„œ í˜ì´ì§€ë‹¹ ì¹´ë“œ ìˆ˜ ì¸¡ì • ===
        thumbs_first = collect_thumb_imgs(driver)
        per_page = max(1, len(thumbs_first))  # 0 ë°©ì§€

        # === ì´ ê°œìˆ˜ â†’ ì´ í˜ì´ì§€ ê³„ì‚° ===
        total_cnt = get_total_count(driver) or per_page
        max_pages = math.ceil(total_cnt / per_page)
        print(
            f"ğŸ“„ ì´ ê²Œì‹œë¬¼: {total_cnt} / í˜ì´ì§€ë‹¹: {per_page} â†’ ê°ì§€ëœ ì´ í˜ì´ì§€: {max_pages}"
        )

        all_rows: List[VideoRow] = []
        seen_global = set()  # (title, youtube_id)
        prev_sig = get_page_signature(driver)

        for page_no in range(1, max_pages + 1):
            # 1í˜ì´ì§€ëŠ” ì´ë¯¸ ë¡œë”©ë¨. 2í˜ì´ì§€ ì´ìƒ ì´ë™
            if page_no > 1:
                ok = goto_page(driver, wait, page_no, PAGE_PAUSE, prev_sig)
                if not ok:
                    print(
                        f"âš ï¸ í˜ì´ì§€ {page_no} ì´ë™ í´ë°± ì‹¤íŒ¨ â†’ 'ë‹¤ìŒ'ìœ¼ë¡œ ì „ì§„ ì‹œë„ í›„ ê³„ì† ìˆ˜ì§‘"
                    )
                    # ê·¸ë˜ë„ ìˆ˜ì§‘ì€ ê³„ì† (í˜„ì¬ ë³´ì´ëŠ” í˜ì´ì§€ë¥¼ íŒŒì‹±)

            # ====== ì¹´ë“œ íŒŒì‹± ======
            unique_cards = collect_cards_from_thumbs(driver)
            print(
                f"[í˜ì´ì§€ {page_no}] ì¸ë„¤ì¼ {len(collect_thumb_imgs(driver))} â†’ ìœ íš¨ ì¹´ë“œ {len(unique_cards)}"
            )

            rows_this_page: List[VideoRow] = []
            rank = 0
            meta_prefixes = (
                "ì²´ë ¥í•­ëª©",
                "ìš´ë™ë„êµ¬",
                "ì‹ ì²´ë¶€ìœ„",
                "ìš´ë™ëŒ€ìƒ",
                "ì§ˆí™˜",
                "ê´€ì‹¬ë™ì˜ìƒ",
            )

            for card in unique_cards:
                txt = (card.text or "").strip()
                if not txt:
                    continue
                lines = [ln.strip() for ln in txt.splitlines() if ln.strip()]

                title_candidates = [
                    ln
                    for ln in lines
                    if not ln.startswith(meta_prefixes) and "ê´€ì‹¬ë™ì˜ìƒ" not in ln
                ]
                if title_candidates:
                    title = title_candidates[0]
                else:
                    fallbacks = [ln for ln in lines if "ê´€ì‹¬ë™ì˜ìƒ" not in ln]
                    title = max(fallbacks, key=len) if fallbacks else ""

                fitness_category = extract_text_after_prefix(lines, "ì²´ë ¥í•­ëª©")
                tool = extract_text_after_prefix(lines, "ìš´ë™ë„êµ¬")
                body_part = extract_text_after_prefix(lines, "ì‹ ì²´ë¶€ìœ„")
                target = extract_text_after_prefix(lines, "ìš´ë™ëŒ€ìƒ")
                disease = extract_text_after_prefix(lines, "ì§ˆí™˜")

                thumb_url, youtube_id = None, None
                try:
                    img = card.find_element(By.CSS_SELECTOR, "img")
                    thumb_url = img.get_attribute("src") or None
                    if thumb_url:
                        m = YOUTUBE_ID_RE.search(thumb_url)
                        if m:
                            youtube_id = m.group(1)
                except:
                    pass
                youtube_url = (
                    f"https://www.youtube.com/watch?v={youtube_id}"
                    if youtube_id
                    else None
                )

                rank += 1
                row = VideoRow(
                    title=title,
                    fitness_category=fitness_category,
                    tool=tool,
                    body_part=body_part,
                    target=target,
                    disease=disease,
                    youtube_id=youtube_id,
                    youtube_url=youtube_url,
                    thumb_url=thumb_url,
                    page_no=page_no,
                    rank_on_page=rank,
                )

                key = (row.title, row.youtube_id)
                if key in seen_global:
                    continue
                seen_global.add(key)
                rows_this_page.append(row)
                time.sleep(ITEM_PAUSE)

            print(f"[í˜ì´ì§€ {page_no}] íŒŒì‹± {len(rows_this_page)}")
            all_rows.extend(rows_this_page)
            print(f"[í˜ì´ì§€ {page_no}] ëˆ„ì  {len(all_rows)}ê°œ")

            # ë‹¤ìŒ ë£¨í”„ ëŒ€ë¹„ í˜„ì¬ ì„œëª… ê°±ì‹ 
            cur_sig = get_page_signature(driver)
            prev_sig = cur_sig or prev_sig

            # ì„ì‹œ ì €ì¥
            if page_no % PARTIAL_SAVE_EVERY == 0:
                df_partial = pd.DataFrame([asdict(r) for r in all_rows])
                df_partial.sort_values(["page_no", "rank_on_page"], inplace=True)
                out_name = f"kspo_partial_{page_no:03d}.csv"
                df_partial.to_csv(out_name, index=False)
                print(f"ğŸ’¾ [ì„ì‹œì €ì¥] {out_name} ({len(df_partial)} rows)")

        # ìµœì¢… ì €ì¥
        df = pd.DataFrame([asdict(r) for r in all_rows])
        df.sort_values(["page_no", "rank_on_page"], inplace=True)
        df.to_csv("kspo_prescription_videos.csv", index=False)
        df.to_json("kspo_prescription_videos.json", orient="records", force_ascii=False)
        print(f"[OK] saved {len(df)} rows to CSV/JSON")

        elapsed = time.time() - t0
        print(f"â±ï¸ ì´ ì†Œìš”ì‹œê°„: {human_seconds(elapsed)}")
        return all_rows

    finally:
        driver.quit()


if __name__ == "__main__":
    try:
        rows = scrape()

        print("\n===== í¬ë¡¤ë§ ì™„ë£Œ ìš”ì•½ =====")
        try:
            df = pd.read_csv("kspo_prescription_videos.csv")
        except Exception:
            df = pd.read_csv("kspo_prescription_videos.csv", encoding="utf-8")

        total = len(df)
        uniq_y = df["youtube_id"].nunique(dropna=True)
        dup_y_cnt = df["youtube_id"].duplicated(keep=False).sum()
        empty_titles = (df["title"].astype(str).str.strip() == "").sum()
        null_y = df["youtube_id"].isna().sum()

        print(f"ì´ ìˆ˜ì§‘: {total}ê°œ")
        print(f"ìœ ë‹ˆí¬ YouTube ID: {uniq_y}ê°œ / ì¤‘ë³µ í‘œê¸° í–‰ ìˆ˜: {dup_y_cnt}ê°œ")
        print(f"ë¹ˆ ì œëª© ë ˆì½”ë“œ: {empty_titles}ê°œ")
        print(f"youtube_idê°€ ë¹„ì–´ìˆëŠ” ë ˆì½”ë“œ: {null_y}ê°œ")

        cols = [
            c
            for c in [
                "title",
                "fitness_category",
                "tool",
                "body_part",
                "target",
                "youtube_url",
            ]
            if c in df.columns
        ]
        print("\nìƒ˜í”Œ 5ê°œ:")
        print(df[cols].head(5).to_string(index=False))

        if total > 0 and (null_y / total) > 0.2:
            print(
                "\nâš ï¸ youtube_id ëˆ„ë½ ë¹„ìœ¨ì´ ë†’ìŠµë‹ˆë‹¤. ì¸ë„¤ì¼ êµ¬ì¡°ê°€ ë°”ë€Œì—ˆì„ ìˆ˜ ìˆì–´ìš”."
            )
        if empty_titles > 0:
            print("âš ï¸ ë¹ˆ ì œëª© ë ˆì½”ë“œê°€ ìˆìŠµë‹ˆë‹¤. ì¹´ë“œ í…ìŠ¤íŠ¸ ì¶”ì¶œ ë¡œì§ì„ ì¡°ì •í•˜ì„¸ìš”.")

        print("\nê²°ê³¼ íŒŒì¼:")
        print(" - kspo_prescription_videos.csv")
        print(" - kspo_prescription_videos.json")

    except KeyboardInterrupt:
        sys.exit(1)
```
